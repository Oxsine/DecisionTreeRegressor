```python
# import statment
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor
```

```python
# setting a seed
rng = np.random.RandomState(2)
```

```python
# random distribution between 0 and 5
X = np.sort(5 * rng.rand(100, 1), axis = 0)
# sin func from X
y = np.sin(X)
```

```python
mean = 0
stddev = 1
# array of zeros with lengt = y
random_noise = np.zeros_like(y)
# every 5th element has normal distribution around mean and borders in standart deviation
random_noise[::5] = np.random.normal(mean, stddev, random_noise[::5].shape)
# apply noise
y += random_noise
plt.scatter(X, y)
plt.title('sin with statistical outlier')
```

```python
# setting up with dif. max depth
regr_1 = DecisionTreeRegressor(max_depth = 1)
regr_2 = DecisionTreeRegressor(max_depth = 2)
regr_3 = DecisionTreeRegressor(max_depth = 3)
regr_4 = DecisionTreeRegressor(max_depth = 4)
regr_5 = DecisionTreeRegressor(max_depth = 5)

#  give then data
regr_1.fit(X, y)
regr_2.fit(X, y)
regr_3.fit(X, y)
regr_4.fit(X, y)
regr_5.fit(X, y)
```

```python
#  array from 0 to 4.999 with step = 0.01
X_test = np.arange(0, 5.0, 0.01)[:, np.newaxis]

# making predicts
y_1 = regr_1.predict(X_test)
y_2 = regr_2.predict(X_test)
y_3 = regr_3.predict(X_test)
y_4 = regr_4.predict(X_test)
y_5 = regr_5.predict(X_test)
```

```python
fig, axs = plt.subplots(3, 2)

fig.subplots_adjust(left=0.2, bottom=0.2, right=1.0, top=1.0, wspace=0.6, hspace=0.6)

axs[0, 0].scatter(X, y, edgecolor = 'black', c = 'blue', label = 'data')
axs[0, 0].plot(X_test, y_1, color = 'red', label = 'max_depth = 1', linewidth = 3)
axs[0, 0].set_title('max_depth = 1')

axs[0, 1].scatter(X, y, edgecolor = 'black', c = 'blue', label = 'data')
axs[0, 1].plot(X_test, y_2, color = 'red', label = 'max_depth = 2', linewidth = 3)
axs[0, 1].set_title('max_depth = 2')

axs[1, 0].scatter(X, y, edgecolor = 'black', c = 'blue', label = 'data')
axs[1, 0].plot(X_test, y_3, color = 'red', label = 'max_depth = 3', linewidth = 3)
axs[1, 0].set_title('max_depth = 3')

axs[1, 1].scatter(X, y, edgecolor = 'black', c = 'blue', label = 'data')
axs[1, 1].plot(X_test, y_4, color = 'red', label = 'max_depth = 4', linewidth = 3)
axs[1, 1].set_title('max_depth = 4')

axs[2, 0].scatter(X, y, edgecolor = 'black', c = 'blue', label = 'data')
axs[2, 0].plot(X_test, y_5, color = 'red', label = 'max_depth = 5', linewidth = 3)
axs[2, 0].set_title('max_depth = 5')

axs[2, 1].set_visible(False)

fig.set_facecolor('lightgray')
```
